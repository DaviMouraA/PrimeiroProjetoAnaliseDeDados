```python
#Plotagens de Variáveis únicas
sns.countplot(data = df_starclass, x = 'class') #Plotando a quantidade de Galaxias, Estrelas e Quasars do Dataset, ou seja, dos diferentes dados da colunas 'class'
```

```python
df_starclass['class'].value_counts() #Contando exatamente a quantidade de cada classe
```

```python
sns.boxplot(data = df_starclass, x = 'class', y = 'u' )#Boxplot de cada classe e sua magnitude aparente nas observações feitas com o filtro ultravioleta
ax = plt.gca()
new_y_range = (0, 35)
ax.set_ylim(new_y_range)
```

```python
sns.boxplot(data = df_starclass, x = 'class', y = 'g' )#Boxplot de cada classe e sua magnitude aparente nas observações feitas com o filtro verde
ax = plt.gca()
ax = plt.gca()
new_y_range = (0, 35)
ax.set_ylim(new_y_range)
```

```python
sns.boxplot(data = df_starclass, x = 'class', y = 'r' )#Boxplot de cada classe e sua magnitude aparente nas observações feitas com o filtro vermelho
ax = plt.gca()
new_y_range = (0, 35)
ax.set_ylim(new_y_range)
```

```python
sns.boxplot(data = df_starclass, x = 'class', y = 'i' )#Boxplot de cada classe e sua magnitude aparente nas observações feitas com o filtro infravermelho próximo
ax = plt.gca()
new_y_range = (0,35)
ax.set_ylim(new_y_range)
```

```python
#Plotagem de Pares de Variáveis
sns.boxplot(data = df_starclass, x = 'class', y = 'z' )#Boxplot de cada classe e sua magnitude aparente nas observações feitas com o filtro infravermelho
ax = plt.gca()
new_y_range = (0, 35)
ax.set_ylim(new_y_range)
```

```python
sns.histplot(data = df_starclass, x = 'class', y = 'redshift')#Verificando os desvios para o vermelho de cada classe.
ax = plt.gca()
new_y_range = (0,8)
ax.set_ylim(new_y_range)
plt.show()
```

```python
sns.scatterplot(data = df_starclass, x = 'class', y = 'redshift')
ax = plt.gca()
new_y_range = (-1, 8)
ax.set_ylim(new_y_range)
plt.show()
```

```python
sns.boxplot(data = df_starclass, x = 'class', y = 'redshift')
ax = plt.gca()
new_y_range = (-1, 10)
ax.set_ylim(new_y_range)
plt.show()
```

```python
filtros = sns.scatterplot(data = df_starclass, x = 'redshift', y = 'i',hue = 'class')
ax = plt.gca()
new_y_range = (-10, 50)
ax.set_ylim(new_y_range)
plt.show()
#É possível concluir com a análise desses gráficos que a classe que contém os maiores redshifts registrados é a dos Quasares, ou seja, essa classe possui os objetos que estão mais distantes de nós e que estão se expandindo mais rapidamente
```

```python
sns.scatterplot(data = df_starclass, x = 'alpha', y = 'delta', hue = 'class')
#Esse gráfico explora a distribuição dos objetos astronômicos no céu, revela agrupamentos e padrões dos objetos usando as
#coordenadas de ascensão reta (alpha(coordenada angular que especifica a posição leste-oeste de um objeto no céu))
#e declinação (delta(ângulo que especifica a posição norte-sul de um objeto no céu. )).
```

```python
filtros = sns.scatterplot(data = df_starclass, x = 'i', y = 'z',hue = 'class')
ax = plt.gca()
new_y_range = (-10, 50)
ax.set_ylim(new_y_range)
plt.show()
#Esse gráfico relacione as magnitudes nos diferentes filtros i e z com a luminosidade aparente das estrelas. Ele ajuda a entender como diferentes tipos de estrelas emitem luz
#em diferentes partes do espectro eletromagnético
```
